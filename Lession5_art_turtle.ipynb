```python
import random
com_player = random.choice(["가위", "바위", "보"])
print(com_player)
user_player = input("가위/바위/보 중 하나를 입력하세요 ==> ")
count = 7
while count > 0 :
  user_player = input("가위/바위/보 중 하나를 입력하세요 ==>")
  if (com_player == user_player) :
    print(" 비겼다 ! ")
  elif ((com_player == "가위") and (user_player == "바위")) or \
    ((com_player == "바위") and (user_player == "보")) or \
    ((com_player == "보") and (user_player == "가위")) :
    print(" 이겼다 !!! ")
  else :
    print(" 졌다 ㅠㅠ ")
  count = count - 1
  
```

형성평가 - 2 : 소스코드

```python
score = int(input("점수를 입력 ==>"))

if score >= 90 :
  print("A", end='')
else :
  if score >= 80 :
    print("B", end='')
  else :
    if score >= 70 :
      print("C", end='')
    else : 
      if score >= 60 :
        print("D", end='')
      else :
        print("F", end='')
print("학점입니다.")
```
